# Bank Account 

```python
# Create a Bank class
class Bank:
    bank = 'UBL'
    
    # Constructor
    def __init__(self, name, balance=0):
        self.name = name
        self.balance = balance

    # Show account details
    def show_balance(self):
        print("Account Holder:", self.name)
        print("Current Balance:", self.balance)

    # Deposit money
    def deposit(self, amount):
        self.balance = self.balance + amount
        print(amount, "Deposited Successfully!")

    # Withdraw money
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance = self.balance - amount
            print(amount, "Withdrawn Successfully!")
        else:
            print("Insufficient Balance!")


# Create Object
customer1 = Bank("Awais", 5000)

# Show Balance
customer1.show_balance()

# Deposit Money
customer1.deposit(2000)

# Show Updated Balance
customer1.show_balance()

# Withdraw Money
customer1.withdraw(3000)

# Show Final Balance
customer1.show_balance()
```

# Encapsulation  


```python


# Create a Bank class
class Bank:

    bank = "UBL"

    # Constructor
    def __init__(self, name, balance=0):
        self.name = name
        self.__balance = balance

    # Show account details
    def show_balance(self):
        print("Bank:", Bank.bank)
        print("Account Holder:", self.name)
        print("Current Balance:", self.__balance)

    # Deposit money
    def deposit(self, amount):
        self.__balance = self.__balance + amount
        print(amount, "Deposited Successfully!")

    # Withdraw money
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance = self.__balance - amount
            print(amount, "Withdrawn Successfully!")
        else:
            print("Insufficient Balance!")


# Create Object
customer1 = Bank("Awais", 5000)

# Show Balance
customer1.show_balance()

# Deposit Money
customer1.deposit(2000)

# Show Updated Balance
customer1.show_balance()

# Withdraw Money
customer1.withdraw(3000)

# Show Final Balance
customer1.show_balance()
```

# IINHERITANCE 

```python
 # Inheritance is a mechanism that allows one class to use the variables and methods of another class without rewriting the code.
# Parent Class
class HaseebAnimal:

    def speak(self):
        print("Haseeb Animal Speaks")


# Child Class
class Dog(HaseebAnimal):

    def speak(self):
        print("Haseeb Dog Barks Loudly")


# Object of Parent Class
a1 = HaseebAnimal()
a1.speak()

# Object of Child Class
d1 = Dog()
d1.speak()
```

# 1. Single Inheritance

```python
# Parent Class
class Employee:

    def login(self):
        print("Employee Logged In")
# Child Class
class Developer(Employee):

    def write_code(self):
        print("Developer is writing code")
# Object
dev = Developer()
dev.login()
dev.write_code()
```

# Multiple Inheritance

```python
# Parent Class 1
class Camera:

    def take_photo(self):
        print("Photo Captured")
# Parent Class 2
class Phone:

    def call(self):
        print("Calling...")
# Child Class
class SmartPhone(Camera, Phone):

    def browse(self):
        print("Browsing Internet")
# Object
mobile = SmartPhone()
mobile.take_photo()
mobile.call()
mobile.browse()
```

# Multilevel Inheritance

```python
# Grandparent → Parent → Child
# Grand Parent
class Company:

    def company_name(self):
        print("Tech Solutions")
# Parent
class Employee(Company):

    def employee_role(self):
        print("Software Engineer")
# Child
class Developer(Employee):

    def programming(self):
        print("Developing Python Applications")
# Object
dev = Developer()

dev.company_name()
dev.employee_role()
dev.programming()
```

# 4. Hierarchical Inheritance

```python
# Parent Class
class Employee:

    def login(self):
        print("Employee Logged In")
# Child Class 1
class Manager(Employee):

    def manage_team(self):
        print("Managing Team")
# Child Class 2
class Accountant(Employee):

    def calculate_salary(self):
        print("Calculating Salary")
# Objects
manager = Manager()
accountant = Accountant()

manager.login()
manager.manage_team()

accountant.login()
accountant.calculate_salary()
```

# 5. Hybrid Inheritance

```python
# Hybrid = Multiple + Hierarchical Inheritance
# Parent Class
class Employee:

    def login(self):
        print("Employee Logged In")
        # Child Class 1
class Developer(Employee):

    def write_code(self):
        print("Writing Code")
# Child Class 2
class Designer(Employee):

    def create_ui(self):
        print("Designing UI")
# Hybrid Class
class TeamLead(Developer, Designer):

    def manage_project(self):
        print("Managing Project")
# Object
lead = TeamLead()
lead.login()
lead.write_code()
lead.create_ui()
lead.manage_project()
```

# USING GET ,SET , PUBLIC ,PRIVATE  AND PROTECTED METHODS 

```python
class Student:
    subject = "AI"

    def __init__(self, name, age, cell):
        # Public Variable
        self.name = name

        # Protected Variable
        self._age = age

        # Private Variable
        self.__cell = cell

    def get_cell(self):
        print(self.__cell)

    def set_cell(self, ph):
        self.__cell = ph


# Object Creation
std1 = Student("Ali", 23, 4343324)

# Access Class Variable
print(std1.subject)

# Get Cell Number
std1.get_cell()

# Update Cell Number
std1.set_cell(334779)

# Check Updated Number
std1.get_cell()
```

# TYPES OF METHODS IN OOPS 

```python
# Class Definition
class Student:

    # Class Variable
    company = "Corvit"

    # Constructor
    def __init__(self, name):
        self.name = name

    # Instance Method
    def change_name(self, name):
        self.name = name

    # Class Method
    @classmethod
    def change_company(cls, name):
        cls.company = name

    # Static Method
    @staticmethod
    def fee(total, bonus_percent):
        fees = total - (bonus_percent * total / 100)
        print(fees)
        print(f"Fee after discount: {fees}")
        return fees


# Object Creation
std1 = Student("Ali")

# Instance Method
std1.change_name("Haseeb")

# Class Method
std1.change_company("Corvit RWP")

# Static Method
Student.fee(5000, 6)

# Check Values
print(std1.name)
print(Student.company)
```
