```python
# Artificial Intelligence (AI) is a branch of Computer Science that enables machines and computers to perform tasks that normally require human intelligence.

# These tasks include:

# Learning from data
# Solving problems
# Making decisions
# Understanding language
# Recognizing images and speech
# Predicting outcomes\
```

```python
from IPython.display import display, HTML

display(HTML("<h1 style='text-align:center; color:#124987861;'>Types of Artificial Intelligence (AI)</h1>"))
```

```python
 # Artificial Intelligence can be classified into two main types:

# 1. Based on Capabilities
# 2. Based on Functionality


# ==========================================
# 1. Based on Capabilities
# ==========================================

# 1. Narrow AI (Weak AI)
# - Performs only one specific task.
# - Cannot perform tasks outside its domain.
# Examples:
# Siri, Alexa, ChatGPT, Google Translate

# 2. General AI (Strong AI)
# - Can perform any task like a human.
# - Can learn, think, and solve different problems.
# Current Status:
# Still under research.

# 3. Super AI
# - More intelligent than humans.
# - Can make better decisions than humans.
# Current Status:
# Future concept (does not exist yet).


# ==========================================
# 2. Based on Functionality
# ==========================================

# 1. Reactive Machines
# - React only to the current situation.
# - Have no memory.
# Example:
# IBM Deep Blue Chess AI

# 2. Limited Memory AI
# - Learns from past data.
# - Most modern AI systems belong to this type.
# Examples:
# Self-driving Cars, ChatGPT, Face Recognition

# 3. Theory of Mind AI
# - Can understand human emotions and behavior.
# Current Status:
# Research stage.

# 4. Self-Aware AI
# - Has self-awareness and consciousness.
# - Can think independently.
# Current Status:
# Science fiction (does not exist yet).
```
