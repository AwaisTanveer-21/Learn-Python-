# Unit Converter App

```python
import tkinter as tk
from tkinter import messagebox

# ---------------- Function ----------------
def convert():
    try:
        value = float(entry.get())
        selected = unit.get()

        if selected == "Meter":
            result = f"""📏 {value} Meter

Feet    : {value*3.28084:.2f}
Inches : {value*39.3701:.2f}"""

        elif selected == "Feet":
            result = f"""👣 {value} Feet

Meter  : {value*0.3048:.2f}
Inches : {value*12:.2f}"""

        else:
            result = f"""📐 {value} Inches

Meter : {value*0.0254:.2f}
Feet     : {value/12:.2f}"""

        output.config(text=result)

    except:
        messagebox.showerror("Error", "Please enter a valid number!")

# ---------------- Window ----------------
root = tk.Tk()
root.title("Unit Converter")
root.geometry("550x500")
root.configure(bg="#EAF4FF")
root.resizable(False, False)

# ---------------- Header ----------------
header = tk.Label(
    root,
    text="UNIT CONVERTER",
    font=("Segoe UI",24,"bold"),
    bg="#2563EB",
    fg="white",
    pady=15
)
header.pack(fill="x")

# ---------------- Card ----------------
card = tk.Frame(
    root,
    bg="white",
    bd=2,
    relief="solid"
)
card.place(x=45,y=90,width=460,height=360)

tk.Label(
    card,
    text="Enter Value",
    font=("Segoe UI",12,"bold"),
    bg="white",
    fg="#374151"
).pack(pady=(20,5))

entry = tk.Entry(
    card,
    font=("Segoe UI",14),
    justify="center",
    bd=2
)
entry.pack(ipady=5,padx=40,fill="x")

tk.Label(
    card,
    text="Select Unit",
    font=("Segoe UI",12,"bold"),
    bg="white",
    fg="#374151"
).pack(pady=15)

unit = tk.StringVar(value="Meter")

menu = tk.OptionMenu(card,unit,"Meter","Feet","Inches")
menu.config(
    font=("Segoe UI",12),
    bg="#3B82F6",
    fg="white",
    width=15
)
menu.pack()

tk.Button(
    card,
    text="Convert",
    font=("Segoe UI",13,"bold"),
    bg="#10B981",
    fg="white",
    activebackground="#059669",
    cursor="hand2",
    command=convert
).pack(pady=20,ipadx=30,ipady=5)

output = tk.Label(
    card,
    text="Result will appear here",
    font=("Segoe UI",13),
    bg="#F3F4F6",
    fg="#111827",
    width=35,
    height=6,
    relief="groove",
    justify="left"
)
output.pack(padx=20,pady=10)

tk.Label(
    root,
    text="Made with Python Tkinter ❤️",
    bg="#EAF4FF",
    fg="gray",
    font=("Segoe UI",10)
).pack(side="bottom",pady=10)

root.mainloop()
```

```python
password = input("Enter your password: ")

upper = False
lower = False
digit = False
special = False

special_characters = "!@#$%^&*()-_=+[]{}|\\:;\"'<>,.?/"

if len(password) >= 8:

    for char in password:

        if char.isupper():
            upper = True

        elif char.islower():
            lower = True

        elif char.isdigit():
            digit = True

        elif char in special_characters:
            special = True

    if upper and lower and digit and special:
        print("Password is VALID")
    else:
        print("Password is INVALID")
        print("Password must contain:")
        print("- At least one uppercase letter")
        print("- At least one lowercase letter")
        print("- At least one digit")
        print("- At least one special character")

else:
    
    print("Password is INVALID")
    print("Password must be at least 8 characters long.")

   
```

## optimal solution for two sum case


```python
def two_sum(nums, target):

    seen = {}

    for i in range(len(nums)):

        complement = target - nums[i]

        if complement in seen:
            return [seen[complement], i]

        seen[nums[i]] = i

    return []


```

```python
nums = [i for i in range(1, 10000001)]

target = 19999999

print(two_sum(nums, target))
```
