- Q1. Differentiate between a Class and an Object with a Real-World Example?

### Class

- A class is a blueprint or template used to create objects. It defines the attributes (variables) and behaviors (methods) that the objects will have.

### Object

An object is a real instance of a class. It contains actual values and can perform the actions defined in the class.

### Real-World Example

Imagine a Car.

**Class:** Car (Blueprint)
**Objects:** Honda Civic, Toyota Corolla, BMW X5

The class defines common properties like:

- Color
- Model
- Speed

Each object has its own values.
