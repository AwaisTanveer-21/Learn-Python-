# Find Mean , Medium and Mode of following 

```python
# Data
age = [22, 24, 21, 23, 25]
salary = [50000, 45000, 48000, 52000, 70000]
gender = ["Male", "Male", "Female", "Male", "Female"]

# ----------------- AGE -----------------

# Mean
total = 0
for i in age:
    total += i
mean_age = total / len(age)

# Median
age.sort()
median_age = age[len(age)//2]

# Mode
count = {}
for i in age:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1

max_count = max(count.values())
mode_age = []
for key in count:
    if count[key] == max_count:
        mode_age.append(key)

# ----------------- SALARY -----------------

# Mean
total = 0
for i in salary:
    total += i
mean_salary = total / len(salary)

# Median
salary.sort()
median_salary = salary[len(salary)//2]

# Mode
count = {}
for i in salary:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1

max_count = max(count.values())
mode_salary = []
for key in count:
    if count[key] == max_count:
        mode_salary.append(key)

# ----------------- GENDER MODE -----------------

count = {}
for i in gender:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1

max_count = max(count.values())
mode_gender = ""

for key in count:
    if count[key] == max_count:
        mode_gender = key

# ----------------- OUTPUT -----------------

print("Mean Age =", mean_age)
print("Median Age =", median_age)
print("Mode Age =", mode_age)

print()

print("Mean Salary =", mean_salary)
print("Median Salary =", median_salary)
print("Mode Salary =", mode_salary)

print()

print("Mode Gender =", mode_gender)
```
